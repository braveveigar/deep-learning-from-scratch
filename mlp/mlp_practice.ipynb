```python
import matplotlib.pyplot as plt
import numpy as np
from functions import *
```

```python
X_train, y_train, X_test, y_test = load_mnist("./dataset")

print("=" * 50)
print("MNIST Dataset Summary")
print("=" * 50)
print(f"Train Images : {X_train.shape}")
print(f"Train Labels : {y_train.shape}")
print(f"Test Images  : {X_test.shape}")
print(f"Test Labels  : {y_test.shape}")
print()

print("Image Info")
print(f"  dtype : {X_train.dtype}")
print(f"  range : [{X_train.min()}, {X_train.max()}]")
print(f"  image size : {X_train.shape[1]} x {X_train.shape[2]}")
print()

print("Labels")
print(f"  classes : {np.unique(y_train)}")
print(f"  samples per class:")
values, counts = np.unique(y_train, return_counts=True)
for v, c in zip(values, counts):
    print(f"    {v} : {c:5d}")
print("=" * 50)

fig, axes = plt.subplots(1, 3, figsize=(10, 3.5))

for ax, idx in zip(axes, [0, 1, 2]):
    ax.imshow(X_train[idx], cmap="gray")
    ax.set_title(f"Label : {y_train[idx]}", fontsize=12, fontweight="bold")
    ax.set_xticks([])
    ax.set_yticks([])

fig.suptitle("MNIST Sample Images", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```
